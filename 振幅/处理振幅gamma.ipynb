```python
import numpy as np
import matplotlib.pyplot as plt
amp_0=np.loadtxt(r"振幅曲线.txt")
amp_1=np.loadtxt(r"振幅曲线-线性-v1.txt")
amp_2=np.loadtxt(r"振幅曲线-线性-v2.txt")

amp_2=amp_2**(1/2)

plt.figure(figsize=(24,6))
plt.subplot(1,3,1)
plt.title("Origin")
plt.plot(amp_0)

plt.subplot(1,3,2)
plt.title("Linnear-1")
plt.plot(amp_1)

plt.subplot(1,3,3)
plt.title("Linnear-2")
plt.plot(amp_2)

plt.show()
```

```python
n=(amp_2-amp_2.min())/(amp_2.max()-amp_2.min())
```

```python
n[:n.argmax()+1].max()
n.argmin()
```

```python
gonglv=n[n.argmin():n.argmax()+1]
pre_gamma=np.arange(n.argmin(),n.argmax()+1)*4
plt.plot(pre_gamma)
pre_gamma[0]
```

```python
# 目标插值后的长度
target_length = 1024

# 生成插值后的索引
indices = np.linspace(0, max(gonglv), target_length)
gamma = np.interp(indices, gonglv, pre_gamma).round()
plt.plot(gamma)
gamma[0]
```

```python
final=np.concatenate([gamma,gamma,gamma])
final.shape

plt.figure()
plt.plot(gamma)
plt.plot(np.loadtxt('671nm振幅型gamma.txt')[0:1024])
plt.show()
```
